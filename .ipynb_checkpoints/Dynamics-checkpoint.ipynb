```python
import numpy as np 
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
from scipy import sparse
import random
from scipy.integrate import solve_ivp
```

```python
dft = pd.read_excel('data.xlsx', sheet_name='TemperatureMatrix')
Tmatrix = dft.iloc[:,:]
T = pd.DataFrame(Tmatrix).to_numpy()
T = np.int_(T)
```

```python
betanonorm = T
Ni = betanonorm.shape[0]
Nj = betanonorm.shape[1]
rp = []
cp = []
P0 = []
ra = []
ca = []
A0 = []
F0 = []
a  = []
G  = []
w  = []
norm0 = []
sigmap = np.zeros([Ni,Nj])
sigmaa = np.zeros([Ni,Nj])
beta = np.zeros([Ni,Nj])
for l in range(Ni):
    rp.append(random.uniform(0, 0.1))
    cp.append(random.uniform(0, 0.1))
    w.append(random.uniform(0, 0.1))
    P0.append(random.uniform(0, 0.1))
    F0.append(random.uniform(0, 1))
    a.append( random.betavariate(2, 5))
for l in range(Nj):
    ra.append(random.uniform(0, 0.1))
    ca.append(random.uniform(0, 0.1))
    A0.append(random.uniform(0, 0.1))
    G.append(random.uniform(1, 2))
for j in range(Nj):
    sumprev = []
    for i in range(Ni):
        sigmap[i][j] = random.uniform(0, 0.1)
        sigmaa[i][j] = random.uniform(0, 0.1)
        sumprev.append(betanonorm[i][j])
        resultpar = sum(sumprev)
    norm0.append(resultpar)


for i in range(Ni):
    for j in range(Nj):
        beta[i][j] = (betanonorm[i][j])/(norm0[j])
P  = np.zeros(Ni)
A  = np.zeros(Nj)
x  = np.concatenate([P, A], axis = 0)
x0 = np.concatenate([P0, A0], axis = 0)
```

```python
def makebinarymatrix(MatrixtoBin):
    BiMatrix = np.zeros([np.int_(MatrixtoBin.shape[0]),np.int_(MatrixtoBin.shape[1])])
    for i in range(MatrixtoBin.shape[0]):
        for j in range(MatrixtoBin.shape[1]):
            if MatrixtoBin[i][j] >= 1:
                BiMatrix[i][j] = 1
    return (BiMatrix)
```

```python
Abin = makebinarymatrix(A)
```

```python
def bmatrix(a):
    """Returns a LaTeX bmatrix

    :a: numpy array
    :returns: LaTeX bmatrix as a string
    """
    if len(a.shape) > 2:
        raise ValueError('bmatrix can at most display two dimensions')
    lines = str(a).replace('[', '').replace(']', '').splitlines()
    rv = [r'\begin{bmatrix}']
    rv += ['  ' + ' & '.join(l.split()) + r'\\' for l in lines]
    rv +=  [r'\end{bmatrix}']
    return '\n'.join(rv)
```

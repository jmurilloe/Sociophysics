```python
import numpy as np
import matplotlib.pyplot as plt

mint, maxt, nt, minx, maxx, nx = 0, 10, 14, 0, 2, 12

t = np.linspace(mint,maxt,nt)
x = np.linspace(minx,maxx,nx)
T, X = np.meshgrid(t,x)

dx = X*(1-X)   #esta es f(x)
dt = np.ones(dx.shape)

dxu = dx/np.sqrt(dt**2+dx**2)
dtu = dt/np.sqrt(dt**2+dx**2)

plt.quiver(T, X, dtu, dxu, color = 'purple', headlength = 0, headwidth = 1)
plt.show
```
